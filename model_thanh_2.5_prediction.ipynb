## Import necessary libraries

```python
import pandas as pd
import sklearn
from sklearn.preprocessing import Normalizer
import numpy as np
```

## Import the dataset

```python
df = pd.read_csv('EPL_2005_2023.csv', index_col=None, header=0)
```

## Preprocessing the data for 2.5 goals prediction

```python
data = df[['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC']]
# data = data[5400:]
data.head()
```

## Add more data based on past variables

```python
# Get the statistics for the last 4 matches of the same team
total_goals_home_4 = []
total_goals_away_4 = []
total_goals_home_got_4 = []
total_goals_away_got_4 = []
total_AS_4 = []
total_AS_got_4 = []
total_HS_4 = []
total_HS_got_4 = []
total_AST_4 = []
total_AST_got_4 = []
total_HST_4 = []
total_HST_got_4 = []
total_AC_4 = []
total_AC_got_4 = []
total_HC_4 = []
total_HC_got_4 = [] 
for i in range(0, len(df)):
    total_goals_home = 0
    total_goals_away = 0
    total_goals_home_got = 0
    total_goals_away_got = 0
    total_AS = 0
    total_AS_got = 0
    total_HS = 0
    total_HS_got = 0
    total_AST = 0
    total_AST_got = 0
    total_HST = 0
    total_HST_got = 0
    total_AC = 0
    total_AC_got = 0
    total_HC = 0
    total_HC_got = 0

    counter_home = 0
    counter_away = 0
    for j in range(i - 1, 0, -1):
        if counter_away < 4:
            if df.loc[i]['AwayTeam'] == df.loc[j]['HomeTeam']:
                counter_away += 1
                total_goals_away += df.loc[j]['FTHG']
                total_goals_away_got += df.loc[j]['FTAG']
                total_AS += df.loc[j]['AS']
                total_AS_got += df.loc[j]['HS']
                total_AST += df.loc[j]['HST']
                total_AST_got += df.loc[j]['AST']
                total_AC += df.loc[j]['HC']
                total_AC_got += df.loc[j]['AC']
            if df.loc[i]['AwayTeam'] == df.loc[j]['AwayTeam']:
                counter_away += 1
                total_goals_away += df.loc[j]['FTAG']
                total_goals_away_got += df.loc[j]['FTHG']
                total_AS += df.loc[j]['HS']
                total_AS_got += df.loc[j]['AS']
                total_AST += df.loc[j]['AST']
                total_AST_got += df.loc[j]['HST']
                total_AC += df.loc[j]['AC']
                total_AC_got += df.loc[j]['HC']
        if counter_home < 4:
            if df.loc[i]['HomeTeam'] == df.loc[j]['HomeTeam']:
                counter_home += 1
                total_goals_home += df.loc[j]['FTHG']
                total_goals_home_got += df.loc[j]['FTAG']
                total_HS += df.loc[j]['AS']
                total_HS_got += df.loc[j]['HS']
                total_HST += df.loc[j]['HST']
                total_HST_got += df.loc[j]['AST']
                total_HC += df.loc[j]['HC']
                total_HC_got += df.loc[j]['AC']
            if df.loc[i]['HomeTeam'] == df.loc[j]['AwayTeam']:
                counter_home += 1
                total_goals_home += df.loc[j]['FTAG']
                total_goals_home_got += df.loc[j]['FTHG']
                total_HS += df.loc[j]['HS']
                total_HS_got += df.loc[j]['AS']
                total_HST += df.loc[j]['AST']
                total_HST_got += df.loc[j]['HST']
                total_HC += df.loc[j]['AC']
                total_HC_got += df.loc[j]['HC']
        if counter_home == 4 and counter_away == 4:
            break
    total_goals_home_4.append(total_goals_home)
    total_goals_away_4.append(total_goals_away)
    total_goals_home_got_4.append(total_goals_home_got)
    total_goals_away_got_4.append(total_goals_away_got)
    total_HS_4.append(total_HS)
    total_HS_got_4.append(total_HS_got)
    total_AS_4.append(total_AS)
    total_AS_got_4.append(total_AS_got)
    total_HST_4.append(total_HST)
    total_HST_got_4.append(total_HST_got)
    total_AST_4.append(total_AST)
    total_AST_got_4.append(total_AST_got)
    total_HC_4.append(total_HC)
    total_HC_got_4.append(total_HC_got)
    total_AC_4.append(total_AC)
    total_AC_got_4.append(total_AC_got)
```

```python
# Append it into the dataset and save the results to a separate CSV for 2.5 predictions.
data['total_goals_home_4'] = total_goals_home_4
data['total_goals_away_4'] = total_goals_away_4
data['total_goals_home_got_4'] = total_goals_home_got_4
data['total_goals_away_got_4'] = total_goals_away_got_4

data['total_AS_4'] = total_AS_4
data['total_AS_got_4'] = total_AS_got_4
data['total_HS_4'] = total_HS_4
data['total_HS_got_4'] = total_HS_got_4

data['total_AST_4'] = total_AST_4
data['total_AST_got_4'] = total_AST_got_4
data['total_HST_4'] = total_HST_4
data['total_HST_got_4'] = total_HST_got_4

data['total_AC_4'] = total_AC_4
data['total_AC_got_4'] = total_AC_got_4
data['total_HC_4'] = total_HC_4
data['total_HC_got_4'] = total_HC_got_4

data.to_csv('EPL_data_for_2.5_prediction_no_bet.csv')
```

## Splitting data into training and validation sets.

```python
results = data['FTHG'] + data['FTAG']
results = [0 if x < 2.5 else 1 for x in results]
features = data.iloc[:, 11:]

training_size = int(len(results) * 0.8)
features = features.interpolate(method='linear')
print(features)
normal = Normalizer()
features = normal.fit_transform(features)
X_train, X_val = np.array(features)[:training_size], np.array(features)[training_size:]
y_train, y_val = np.array(results)[:training_size], np.array(results)[training_size:]
```

## Perform the classification

```python
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
print(y_pred)
print(accuracy_score(model.predict(X_train), y_train))
print(accuracy_score(y_pred, y_val))
```
