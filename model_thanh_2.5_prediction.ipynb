## Import necessary libraries

## Import the dataset

```python
df = pd.read_csv('EPL_2005_2023.csv', index_col=None, header=0)
```

## Preprocessing the data for 2.5 goals prediction

```python
data = df[['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'HC', 'AC', 'B365>2.5', 'B365<2.5', 'P>2.5', 'P<2.5']]
data = data[5400:]
data.head()
```

## Add more data based on past variables

```python
# Get the statistics for the last 4 matches of the
total_goals_home_4 = []
total_goals_away_4 = []
total_goals_home_got_4 = []
total_goals_away_got_4 = []
total_AS_4 = []
total_AS_got_4 = []
total_HS_4 = []
total_HS_got_4 = []
total_AST_4 = []
total_AST_got_4 = []
total_HST_4 = []
total_HST_got_4 = []
total_AC_4 = []
total_AC_got_4 = []
total_HC_4 = []
total_HC_got_4 = [] 
for i in range(5400, len(df)):
    total_goals_home = 0
    total_goals_away = 0
    total_goals_home_got = 0
    total_goals_away_got = 0
    total_AS = 0
    total_AS_got = 0
    total_HS = 0
    total_HS_got = 0
    total_AST = 0
    total_AST_got = 0
    total_HST = 0
    total_HST_got = 0
    total_AC = 0
    total_AC_got = 0
    total_HC = 0
    total_HC_got = 0

    counter_home = 0
    counter_away = 0
    for j in range(i - 1, 0, -1):
        if counter_away < 4:
            if df[i]['AwayTeam'] == df[j]['HomeTeam']:
                counter_away += 1
                total_goals_away += df[j]['FTHG']
                total_goals_away_got += df[j]['FTAG']
                total_AS += df[j]['AS']
                total_AS_got += df[j]['HS']
                total_AST += df[j]['HST']
                total_AST_got += df[j]['AST']
                total_AC += df[j]['HC']
                total_AC_got += df[j]['AC']
            if df[i]['AwayTeam'] == df[j]['AwayTeam']:
                counter_away += 1
                total_goals_away += df[j]['FTAG']
                total_goals_away_got += df[j]['FTHG']
                total_AS += df[j]['HS']
                total_AS_got += df[j]['AS']
                total_AST += df[j]['AST']
                total_AST_got += df[j]['HST']
                total_AC += df[j]['AC']
                total_AC_got += df[j]['HC']
        if counter_home < 4:
            if df[i]['HomeTeam'] == df[j]['HomeTeam']:
                counter_home += 1
                total_goals_home += df[j]['FTHG']
                total_goals_home_got += df[j]['FTAG']
                total_HS += df[j]['AS']
                total_HS_got += df[j]['HS']
                total_HST += df[j]['HST']
                total_HST_got += df[j]['AST']
                total_HC += df[j]['HC']
                total_HC_got += df[j]['AC']
            if df[i]['HomeTeam'] == df[j]['AwayTeam']:
                counter_home += 1
                total_goals_home += df[j]['FTAG']
                total_goals_home_got += df[j]['FTHG']
                total_HS += df[j]['HS']
                total_HS_got += df[j]['AS']
                total_HST += df[j]['AST']
                total_HST_got += df[j]['HST']
                total_HC += df[j]['AC']
                total_HC_got += df[j]['HC']
        if counter_home == 4 and counter_away == 4:
            total_goals_home_4.append(total_goals_home)
            total_goals_away_4.append(total_goals_away)
            total_goals_home_got_4.append(total_goals_home_got)
            total_goals_away_got_4.append(total_goals_away_got)
            total_HS_4.append(total_HS)
            total_HS_got_4.append(total_HS_got)
            total_AS_4.append(total_AS)
            total_AS_got_4.append(total_AS_got)
            total_HST_4.append(total_HST)
            total_HST_got_4.append(total_HST_got)
            total_AST_4.append(total_AST)
            total_AST_got_4.append(total_AST_got)
            total_HC_4.append(total_HC)
            total_HC_got_4.append(total_HC_got)
            total_AC_4.append(total_AC)
            total_AC_got_4.append(total_AC_got)
            
            break
```
